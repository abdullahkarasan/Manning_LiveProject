# Partial Solution- Calculating Return and Daily Risk Free Rate

```python
# Note: The data below is the one you retrieved in previous part and it is used here to run the rest of the analysis.
import pandas as pd

stock_prices = pd.read_csv('stock_prices.csv')
stock_prices = stock_prices.set_index('Date')
risk_free =  pd.read_csv('risk_free.csv')
risk_free = risk_free.set_index('Unnamed: 0')
risk_free.index.names = ['Date']
```

```python
returns = stock_prices.pct_change()
returns.dropna(inplace=True)
returns.head()
```

```python
rf = risk_free / 90
rf = rf.dropna().iloc[:-1]
```

```python
stocks = ["AAPL", "IBM", "MSFT", "INTC", "GSPC"]
for i in stocks:
    returns["excess_return_"+str(i)] = returns[i] - rf['0']
```
