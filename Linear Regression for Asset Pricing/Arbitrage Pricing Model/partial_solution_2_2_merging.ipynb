```python
# Note: The data below is the one you retrieved in previous part and
#it is used here to run the rest of the analysis.
import pandas as pd

apt_df = pd.read_csv('apt_df.csv')
apt_df = apt_df.set_index('timestamp')
returns =  pd.read_csv('returns.csv')
returns = returns.set_index('timestamp')
risk_free_reindexed = pd.read_csv('risk_free_reindexed.csv')
risk_free_reindexed = risk_free_reindexed.set_index('timestamp')
risk_free = risk_free_reindexed.copy()
```

```python
cols = returns.columns
for i in cols:
    returns["excess_return_"+str(i)] = returns[i] - risk_free['0']
```

```python
returns.head()
```

```python
returns.to_csv('returns.csv')
```

```python
df_merged = pd.concat([returns.iloc[:, 3:], apt_df.loc[:,['inf_rate', 'log_ind']]],
                      axis=1, ignore_index=True)
df_merged.rename(columns={0:'excess_return_TWTR',
                        1:'excess_return_FB',
                        2:'excess_return_SPY',
                        3:'inf_rate',
                        4:'log_ind'}, inplace=True)
```
